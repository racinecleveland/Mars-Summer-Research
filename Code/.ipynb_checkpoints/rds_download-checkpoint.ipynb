```python
import pandas as pd
import requests
from pathlib import Path

# -----------------------
# settings
# -----------------------
csv_path = "E:/mars2020_meda_calibrated_RDS.csv"
urn_col = "URN"

out_dir = Path("E:/rds_data")
out_dir.mkdir(exist_ok=True)

base_url = "https://atmos.nmsu.edu/PDS/data/PDS4/Mars2020/mars2020_meda/data_calibrated_env"

# -----------------------
# load CSV
# -----------------------
df = pd.read_csv(csv_path)
df.columns = [c.strip() for c in df.columns]
df[urn_col] = df[urn_col].astype(str).str.strip()

# -----------------------
def sol_to_range_folder(sol: int) -> str:
    if sol < 180:
        start = (sol // 90) * 90
        end = start + 89
    else:
        start = 180 + ((sol - 180) // 120) * 120
        end = start + 119
    return f"sol_{start:04d}_{end:04d}"

def urn_to_filename(urn: str) -> str:
    stem = urn.split(":")[-1].upper()
    return f"{stem}01.CSV"

```

```python
# -----------------------
# download loop
# -----------------------
session = requests.Session()

success = 0
failed = []

for urn in df[urn_col].dropna().unique():
    try:
        stem = urn.split(":")[-1].upper()
        sol = int(stem[4:8])

        # --- THIS IS WHERE YOUR NEW LINES GO ---
        folder = sol_to_range_folder(sol)
        sol_folder = f"sol_{sol:04d}"

        found = False

        for suffix in ["01.CSV", "02.CSV", "03.CSV"]:
            fname = f"{stem}{suffix}"
            url = f"{base_url}/{folder}/{sol_folder}/{fname}"

            r = session.get(url, timeout=60)

            if r.status_code == 200:
                with open(out_dir / fname, "wb") as f:
                    f.write(r.content)
            
                success += 1
                found = True
                break

        if not found:
            failed.append(urn)
            print(f"✗ {urn}")

    except Exception as e:
        failed.append((urn, str(e)))
        print(f"✗ {urn}")

print(f"\nDownloaded: {success}")
print(f"Failed: {len(failed)}")
```
