```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 10, 0.1)
y = x**2

fig = plt.figure(figsize = (9,3))

ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y, 'r--')

ax1.set_xlabel('Time')
ax1.set_ylabel('Return', labelpad=5)

fig.suptitle('Example Figure')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
fig.subplots_adjust(top=0.8)
ax1 = fig.add_subplot(211)
ax1.set_ylabel('Voltage [V]')
ax1.set_title('A sine wave')

t = np.arange(0.0, 1.0, 0.01)
s = np.sin(2*np.pi*t)
line, = ax1.plot(t, s, color='blue', lw=2)

# Fixing random state for reproducibility
np.random.seed(19680801)

ax2 = fig.add_axes((0.15, 0.1, 0.7, 0.3))
n, bins, patches = ax2.hist(np.random.randn(1000), 50,
                            facecolor='yellow', edgecolor='yellow')
ax2.set_xlabel('Time [s]')

plt.show()
```

```python
import matplotlib.lines as lines

fig = plt.figure()

l1 = lines.Line2D([0, 1], [0, 1], transform=fig.transFigure, figure=fig, color='red', linewidth=20.0)
l2 = lines.Line2D([0, 1], [1, 0], transform=fig.transFigure, figure=fig, color='red', linewidth=20.0)
fig.lines.extend([l1, l2])

ax = fig.add_subplot()

rect = ax.patch  # a Rectangle instance
rect.set_facecolor('blue')


plt.show()
```
