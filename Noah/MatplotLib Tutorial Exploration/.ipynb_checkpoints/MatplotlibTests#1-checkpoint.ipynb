```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
#matplotlib assumes the values to be on the y-scale, and having not been given x values, it auto generates them

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
#here, however, we gave it both, but didn't clear the first set, thereby drawing on both lines

plt.xlabel('time')
plt.ylabel('some numbers')
plt.show()


plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
#this third argument are contacted both the color 'r' and the shape 'o'
# 'b-' for example, would be a blue line

plt.axis((0, 6, 0, 20))
plt.show()

plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-')
#this third argument are contacted both the color 'r' and the shape 'o'
# 'b-' for example, would be a blue line

plt.axis((0, 6, 0, 20))
#x min, x max, y min, y max
plt.show()
```

```python
import numpy as np

# evenly sampled time at 200ms intervals
t = np.arange(0., 5., 0.2)

# red dashes, blue squares, green triangles, and a yellow line
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^', t, t**3+50, 'y-')
plt.show()
```

```python
'''This creates a dictionary with the keys of 'a' 'c' and 'd' 
    with the related values of three arrays of 50 elements, first, under 'a' an array of integers
    on the range from 0 to 49 (50 is exclusive). Next, 'c' are random integers from 0 to 49.
    Finally, 'd' are 50 numbers pulled using normal distribution
'''
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}

'''element under key 'b' is created from the set of data 'a', added with a random number
in the normal distribution times ten'''
data['b'] = data['a'] + 10 * np.random.randn(50)

#this just normalizes the values of d, which were on a random normal distribution, and multiplies them by 100
data['d'] = np.abs(data['d']) * 100

#Create a scatter plot with the data we've just created.
#The argument's we're using are the x values as 'a', the y values as 'b', 'c' 
#are the colors and data is what we're filling in
plt.scatter('a', 'b', c='c', s='d', data=data)

#Set labels and show
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()


#essentially, we're supplying a dicitonary chocked full of data and we're also supplying the keys to pull off it
```

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

#create the figure
plt.figure(figsize=(9, 3))

'''add a subplot, specifying the number of rows, number of columns, and index working on
  the same as just using plt.subplot(1,3,1) '''
plt.subplot(131)
#add a bar graph on this subplot
plt.bar(names, values)
plt.subplot(132)
#add a scatter plot on this subplot
plt.scatter(names, values)
plt.subplot(133)
#add a line graph on this subplot
plt.plot(names, values)
#add a title over the entire thing
plt.suptitle('Categorical Plotting')
#Show the plot we've worked on!
plt.show()
```

```python
#quick shape function
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

#get us some data. t1 is just more frequent than t2
t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

#create the figure
plt.figure() #the figure call is unnessecary here. If there isn't one it will be created
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

#new plot
plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')

#display them graphs\
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot() by default

plt.figure(1)                # first figure current;
                             # subplot(212) still current
plt.subplot(211)             # make subplot(211) in the first figure
                             # current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)


plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')

#arbitrarily add text onto the image
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

```python
from matplotlib.font_manager import FontProperties

font = FontProperties(family='Garamond', style='italic', weight='bold')

x1 = np.linspace(0.0, 5.0, 100)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)

fig, ax = plt.subplots(figsize=(5, 3))
fig.subplots_adjust(bottom=0.15, left=0.2)
ax.plot(x1, y1*10000)
ax.set_xlabel('Time (s)', fontproperties=font)
ax.set_ylabel('Damped oscillation (V) \n mehhhhhhhhh', labelpad=-500, horizontalalignment='left', fontproperties=font)

plt.show()
```

```python
fig, axs = plt.subplots(3, 1, figsize=(5, 6), tight_layout=True)
locs = ['center', 'left', 'right']
for ax, loc in zip(axs, locs):
    ax.plot(x1, y1)
    ax.set_title('Title with loc at ' + loc, loc=loc)
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
axs[1].xaxis.set_ticks(np.arange(0., 8.1, 1.))
plt.show()
```

```python
fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
ticks = np.arange(0., 8.1, 1.)
# list comprehension to get all tick labels...
tickla = [f'{tick:1.5f}' for tick in ticks]
axs[1].xaxis.set_ticks(ticks)
axs[1].xaxis.set_ticklabels(tickla)
axs[1].set_xlim(axs[0].get_xlim())
plt.show()
```

```python
import matplotlib

fig, axs = plt.subplots(2, 1, figsize=(5, 3), tight_layout=True)
axs[0].plot(x1, y1)
axs[1].plot(x1, y1)
locator = matplotlib.ticker.FixedLocator(ticks)
axs[1].xaxis.set_major_locator(locator)
axs[1].xaxis.set_major_formatter('±{x}°')
plt.show()
```

```python
import datetime
import matplotlib.dates as mdates

locator = mdates.DayLocator(bymonthday=[1, 15])
formatter = mdates.DateFormatter('%b %d')

fig, ax = plt.subplots(figsize=(5, 3), tight_layout=True)
ax.xaxis.set_major_locator(locator)
ax.xaxis.set_major_formatter(formatter)
base = datetime.datetime(2017, 1, 1, 0, 0, 1)
time = [base + datetime.timedelta(days=x) for x in range(len(x1))]

ax.plot(time, y1)
ax.tick_params(axis='x', rotation=70)
plt.show()
```

```python
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```
